```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the style for seaborn plots
sns.set(style="whitegrid")
```

```python
df = pd.read_csv(r'SampleSuperstore.csv')
print(df.head())
```

```python
# Basic information about the dataset
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Descriptive statistics
print(df.describe())
```

```python

plt.figure(figsize=(10,6))
sns.histplot(df['Profit'], kde=True, bins=30, color='orange')
plt.title('Distribution of Profit')
plt.show()
```

```python
print('Basic information ablout datatypes and cloumn values')
df.info()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Profit'], kde=True, bins=30, color='orange')
plt.title('Distribution of Profit')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Sales', y='Profit', hue='Region', data=df)
plt.title('Sales vs. Profit by Region')
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Category', y='Sales', data=df)
plt.title('Sales by Category')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Sub-Category', y='Profit', data=df)
plt.title('Profit by Sub-Category')
plt.xticks(rotation=90)
plt.show()
```
