```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Replace 'path_to_dataset.csv' with the actual path to your downloaded dataset
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.isnull().sum()  # Check for missing values

# Drop rows with missing values (example)
df.dropna(inplace=True)

# Alternatively, fill missing values (example)
# df.fillna(method='ffill', inplace=True)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df['iyear'].value_counts().sort_index().plot(kind='bar', figsize=(12, 6))
plt.title('Number of Terrorist Attacks per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
top_countries = df['country_txt'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries with the Most Terrorist Attacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.show()
```

```python
# Example: Number of attacks by region
top_regions = df['region_txt'].value_counts().head(10)
sns.barplot(x=top_regions.values, y=top_regions.index, palette='coolwarm')
plt.title('Top Regions with the Most Terrorist Attacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Region')
plt.show()
```

```python
attack_types = df['attacktype1_txt'].value_counts().head(10)
sns.barplot(x=attack_types.values, y=attack_types.index, palette='inferno')
plt.title('Top Types of Terrorist Attacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Attack Type')
plt.show()
```

```python
target_types = df['targtype1_txt'].value_counts().head(10)
sns.barplot(x=target_types.values, y=target_types.index, palette='magma')
plt.title('Top Targets of Terrorist Attacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Target Type')
plt.show()
```

```python
weapon_types = df['weaptype1_txt'].value_counts().head(10)
sns.barplot(x=weapon_types.values, y=weapon_types.index, palette='plasma')
plt.title('Top Weapons Used in Terrorist Attacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Weapon Type')
plt.show()
```
